```python
import pandas as pd
import numpy as np    
from datetime import datetime
```

```python
#read csv file 

citibike_data1 = pd.read_csv('/Users/archanakale/Desktop/City Bike /JC-202401-citibike-tripdata.csv')
citibike_data2 = pd.read_csv('/Users/archanakale/Desktop/City Bike /JC-202402-citibike-tripdata.csv')
citibike_data3 = pd.read_csv('/Users/archanakale/Desktop/City Bike /JC-202403-citibike-tripdata.csv')

#concatenate the dataframes
citibike_data = pd.concat([citibike_data1, citibike_data2, citibike_data3])

#check the data
citibike_data.head()

```

```python
# clean the data
citibike_df = citibike_data.dropna()
citibike_df = citibike_df.drop_duplicates()  
citibike_df.head()
```

```python
# create new columns 'distance' to calculate the distance between start and end station
citibike_df['distance'] = np.sqrt((citibike_df['start_lat'] - citibike_df['end_lat'])**2 + (citibike_df['start_lng'] - citibike_df['end_lng'])**2)
citibike_df.head()

  

```

```python
# split the 'started_at' and 'ended_at' columns into date and time
citibike_df['started_at'] = pd.to_datetime(citibike_df['started_at'])
citibike_df['ended_at'] = pd.to_datetime(citibike_df['ended_at'])

citibike_df['start_date'] = citibike_df['started_at'].dt.date
citibike_df['start_time'] = citibike_df['started_at'].dt.time

citibike_df['end_date'] = citibike_df['ended_at'].dt.date
citibike_df['end_time'] = citibike_df['ended_at'].dt.time

citibike_df.head()

```

```python
# create a 'month' and 'Day' column to extract the month name and day from the 'started_at' column
citibike_df['month'] = citibike_df['started_at'].dt.month_name()
citibike_df['day'] = citibike_df['started_at'].dt.day_name()

citibike_df.head()
```

```python
# save the cleaned data to a new excel file
citibike_df.to_excel('/Users/archanakale/Desktop/City Bike /citibike_data.xlsx', index=False)
```
